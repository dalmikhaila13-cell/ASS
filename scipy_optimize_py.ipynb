```python
from scipy.optimize import minimize
import numpy as np

def objective_function(x):
    x_val = x[0]
    y_val = x[1]
    return (x_val - 3)**2 + y_val**2

initial_guess = [0.0, 0.0]

result = minimize(
    fun=objective_function,
    x0=initial_guess,
    method='BFGS'
)

print("--- SciPy Optimization Results ---")
if result.success:
    print("Optimization successful.")
    print(f"Minimum Cost found: {result.fun:.6f}")
    print(f"Optimal parameters (x, y): {result.x}")
else:
    print("Optimization failed.")
```
