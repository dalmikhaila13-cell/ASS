```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

data = {
    'Frequency': np.random.uniform(200, 20000, 100),
    'AngleOfAttack': np.random.uniform(0, 15, 100),
    'ChordLength': np.random.uniform(0.02, 0.3, 100),
    'Velocity': np.random.uniform(30, 70, 100),
    'Thickness': np.random.uniform(0.0004, 0.005, 100),
    'SPL': np.random.normal(120, 5, 100) + (np.random.uniform(-5, 5, 100) * (np.random.rand(100) > 0.5))
}
df = pd.DataFrame(data)

X = df.drop('SPL', axis=1)
y = df['SPL']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model = LinearRegression()
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)
mse = mean_squared_error(y_test, y_pred)

print("--- Airfoil Self-Noise Modeling Results ---")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print("Model trained and evaluated successfully.")
```
