```python
import random

TARGET = "HEART"
TARGET_LEN = len(TARGET)
POPULATION_SIZE = 100
MUTATION_RATE = 0.01

def create_individual():
    return "".join(random.choice("ABCDEFGHIJKLMNOPQRSTUVWXYZ ") for _ in range(TARGET_LEN))

def calculate_fitness(individual):
    score = 0
    for i in range(TARGET_LEN):
        if individual[i] == TARGET[i]:
            score += 1
    return score

def select_parent(population, fitnesses):
    total_fitness = sum(fitnesses)
    selection_point = random.uniform(0, total_fitness)
    current_point = 0
    for individual, fitness in zip(population, fitnesses):
        current_point += fitness
        if current_point >= selection_point:
            return individual
    return population[-1]

def crossover(parent1, parent2):
    split_point = random.randint(1, TARGET_LEN - 1)
    child = parent1[:split_point] + parent2[split_point:]
    return child

def mutate(individual):
    new_individual = list(individual)
    for i in range(TARGET_LEN):
        if random.random() < MUTATION_RATE:
            new_individual[i] = random.choice("ABCDEFGHIJKLMNOPQRSTUVWXYZ ")
    return "".join(new_individual)

def genetic_algorithm():
    population = [create_individual() for _ in range(POPULATION_SIZE)]

    for generation in range(5000):
        fitnesses = [calculate_fitness(ind) for ind in population]
        best_individual = population[fitnesses.index(max(fitnesses))]
        best_fitness = max(fitnesses)

        if best_fitness == TARGET_LEN:
            print(f"Goal reached in Generation {generation}: {best_individual}")
            return

        print(f"Generation {generation}: Best = {best_individual} (Fitness: {best_fitness})")

        new_population = []
        for _ in range(POPULATION_SIZE):
            parent1 = select_parent(population, fitnesses)
            parent2 = select_parent(population, fitnesses)
            child = crossover(parent1, parent2)
            child = mutate(child)
            new_population.append(child)

        population = new_population

genetic_algorithm()
```
