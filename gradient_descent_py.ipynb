```python
def cost_function(theta):
    return (theta - 10)**2

def gradient(theta):
    return 2 * (theta - 10)

learning_rate = 0.01
theta = 0.0
iterations = 1000

print("Starting Gradient Descent...")
for i in range(iterations):
    grad = gradient(theta)
    theta = theta - learning_rate * grad

    if (i + 1) % 200 == 0:
        print(f"Iteration {i+1}: Theta = {theta:.4f}, Cost = {cost_function(theta):.6f}")

print("\n--- Final Result ---")
print(f"Optimal Theta found: {theta:.4f}")
print(f"Minimum Cost achieved: {cost_function(theta):.6f}")
```
