```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
```

```python
N = 100000

points_inside_circle = 0

X_circle = []
Y_circle = []
X_square = []
Y_square = []
```

```python
for i in range(N):
  x = random.random()
  y = random.random()
if x**2 + y**2 <= 1:
  points_inside_circle += 1
  X_circle.append(x)
  Y_circle.append(y)
else:
  X_square.append(x)
  Y_square.append(y)
```

```python
pi_estimate = 4 * (points_inside_circle / N)

print(f"--- Monte Carlo Pi Estimation ---")
print(f"Total points (N): {N}")
print(f"Points inside circle: {points_inside_circle}")
print(f"Estimated Pi: {pi_estimate}")
print(f"Actual Pi (for comparison): {math.pi}")
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(X_square, Y_square, color='red', s=1, label='Outside Circle')

plt.scatter(X_circle, Y_circle, color='blue', s=1, label='Inside Circle')

plt.title(f"Monte Carlo Pi Estimation: {pi_estimate:.4f}")
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
plt.legend()
plt.show()
```

```python
shape_param = 2.0
scale_param = 2.0
population_size = 100000

population_data = np.random.gamma(shape_param, scale_param, population_size)

num_samples = 1000
sample_size = 100

sample_means = []
```

```python
for i in range(num_samples):
    sample = np.random.choice(population_data, size=sample_size, replace=False)

    current_mean = np.mean(sample)

    sample_means.append(current_mean)

print(f"We calculated {len(sample_means)} sample means.")
print(f"The average of all these means is: {np.mean(sample_means):.4f}")
```

```python
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.hist(population_data, bins=50, density=True, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("1. Original Population (Gamma Distribution)")
plt.xlabel("Value")
plt.ylabel("Density")

plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, density=True, color='lightcoral', edgecolor='black', alpha=0.7)
plt.title("2. Distribution of Sample Means (Central Limit Theorem)")
plt.xlabel("Sample Mean")
plt.ylabel("Density")

expected_mean = shape_param * scale_param
plt.axvline(expected_mean, color='darkgreen', linestyle='--', linewidth=2, label=f'Expected Mean: {expected_mean}')
plt.legend()

plt.tight_layout()
plt.show()
```
