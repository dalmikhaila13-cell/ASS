```python
import numpy as np

RULE = 30
SIZE = 50
GENERATIONS = 50

binary_rule = np.binary_repr(RULE, width=8)
RULE_SET = {
    '111': binary_rule[0], '110': binary_rule[1],
    '101': binary_rule[2], '100': binary_rule[3],
    '011': binary_rule[4], '010': binary_rule[5],
    '001': binary_rule[6], '000': binary_rule[7]
}

automaton = np.zeros(SIZE, dtype=int)
automaton[SIZE // 2] = 1

history = [automaton.copy()]

def apply_rule(left, center, right):
    neighborhood = str(left) + str(center) + str(right)
    return int(RULE_SET[neighborhood])

for gen in range(GENERATIONS):
    new_automaton = automaton.copy()

    for i in range(1, SIZE - 1):
        left = automaton[i-1]
        center = automaton[i]
        right = automaton[i+1]

        new_automaton[i] = apply_rule(left, center, right)

    automaton = new_automaton
    history.append(automaton.copy())

print(f"Cellular Automata (Rule {RULE}) over {GENERATIONS} generations:")
for row in history:
    print("".join(['#' if cell == 1 else '.' for cell in row]))
```
