```python
import numpy as np
import matplotlib.pyplot as plt

S0 = 100
T = 1.0
mu = 0.05
sigma = 0.2
N = 252
M = 1000
dt = T / N

nu = mu - (sigma**2) / 2
sqrt_dt = np.sqrt(dt)

Z = np.random.normal(size=(M, N))

exponent_term = (nu * dt) + (sigma * sqrt_dt * Z)

price_paths = np.zeros((M, N + 1))
price_paths[:, 0] = S0
price_paths[:, 1:] = S0 * np.exp(np.cumsum(exponent_term, axis=1))

time = np.linspace(0, T, N + 1)

plt.figure(figsize=(10, 6))
plt.plot(time, price_paths[:10].T)
plt.title(f'Monte Carlo Simulation of Stock Price (M={M} paths)')
plt.xlabel('Time (Years)')
plt.ylabel('Stock Price ($)')
plt.grid(True)
plt.show()

final_prices = price_paths[:, -1]
print(f"Mean final price: ${np.mean(final_prices):.2f}")
print(f"Standard deviation of final price: ${np.std(final_prices):.2f}")
```
