```python
import numpy as np
import matplotlib.pyplot as plt

T = 1.0
N = 1000
dt = T / N

random_steps = np.random.normal(0, np.sqrt(dt), N)

W = np.cumsum(random_steps)

W = np.insert(W, 0, 0)

time = np.linspace(0, T, N + 1)

plt.figure(figsize=(10, 6))
plt.plot(time, W)
plt.title('Standard Brownian Motion (Wiener Process)')
plt.xlabel('Time (t)')
plt.ylabel('W(t)')
plt.grid(True)
plt.show()

print(f"Final value of W(T): {W[-1]:.4f}")
```
